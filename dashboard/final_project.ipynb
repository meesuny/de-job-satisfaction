```python
myquery = '''
    SELECT "AGE" as factor_cat, (sum("JOBSATIS")/count("JOBSATIS")) as Avg_Job_Satisfaction
    FROM db2
    where "JOBSATIS" < 9
    group by "AGE"
'''
age_js = pd.read_sql(myquery,con=engine)
age_js

age_jobsat = pd.DataFrame(age_js, columns=['factor_cat', 'avg_job_satisfaction'])
age_jobsat['factor'] = "age"
age_jobsat['factor_id'] = "F01"
age_jobsat
```

```python
myquery = '''
    SELECT "GENDER" as factor_cat, (sum("JOBSATIS")/count("JOBSATIS")) as Avg_Job_Satisfaction
    FROM db2
    where "JOBSATIS" < 9
    group by "GENDER"
'''
gender_js = pd.read_sql(myquery,con=engine)
gender_js

gender_jobsat = pd.DataFrame(gender_js, columns=['factor_cat', 'avg_job_satisfaction'])
gender_jobsat['factor'] = "gender"
gender_jobsat['factor_id'] = "F02"
gender_jobsat
```

```python
myquery = '''
    SELECT "RACETH" as factor_cat, (sum("JOBSATIS")/count("JOBSATIS")) as Avg_Job_Satisfaction
    FROM db2
    where "JOBSATIS" < 9
    group by "RACETH"
'''
race_js = pd.read_sql(myquery,con=engine)
race_js

race_jobsat = pd.DataFrame(race_js, columns=['factor_cat', 'avg_job_satisfaction'])
race_jobsat['factor'] = "race"
race_jobsat['factor_id'] = "F03"
race_jobsat
```

```python
myquery = '''
    SELECT "DGRDG" as factor_cat, (sum("JOBSATIS")/count("JOBSATIS")) as Avg_Job_Satisfaction
    FROM db2
    where "JOBSATIS" < 9
    group by "DGRDG"
'''
deg_js = pd.read_sql(myquery,con=engine)
deg_js

deg_jobsat = pd.DataFrame(deg_js, columns=['factor_cat', 'avg_job_satisfaction'])
deg_jobsat['factor'] = "education"
deg_jobsat['factor_id'] = "F04"
deg_jobsat
```

```python
myquery = '''
    SELECT "SALARY" as factor_cat, (sum("JOBSATIS")/count("JOBSATIS")) as Avg_Job_Satisfaction
    FROM db2
    where "JOBSATIS" < 9
    group by "SALARY"
'''
salary_js = pd.read_sql(myquery,con=engine)
salary_js

salary_jobsat = pd.DataFrame(salary_js, columns=['factor_cat', 'avg_job_satisfaction'])
salary_jobsat['factor'] = "salary"
salary_jobsat['factor_id'] = "F05"
salary_jobsat
```

```python
myquery = '''
    SELECT "CHTOT" as factor_cat, (sum("JOBSATIS")/count("JOBSATIS")) as Avg_Job_Satisfaction
    FROM db2
    where "JOBSATIS" < 9 and "CHTOT" < 90
    group by "CHTOT"
'''
kids_js = pd.read_sql(myquery,con=engine)
kids_js

kids_jobsat = pd.DataFrame(kids_js, columns=['factor_cat', 'avg_job_satisfaction'])
kids_jobsat['factor'] = "number of children"
kids_jobsat['factor_id'] = "F06"
kids_jobsat
```

## Apply theme

```python
#external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']
external_stylesheets = [dbc.themes.MINTY]
```

## Initialize the app with JupyterDash

```python
app = JupyterDash(__name__, external_stylesheets=external_stylesheets) # Jupyter version
#app = dash.Dash(__name__, external_stylesheets=external_stylesheets) # regular version
```

## Specify contents of dashboard

```python

app.layout = html.Div(
[
    html.H1("Job Satisfaction Dashboard"),
    
    dcc.Dropdown(id = 'factorselect', options = factorlist, value = 'F01'),
    
        html.Div(
            [dcc.Graph(id = 'factorgraph')],
            style = {'width': '65%', 'float':'left'}
        )
]    
)
```

## Callbacks

```python
@app.callback(Output(component_id = 'factorgraph', component_property = 'figure'), 
              Input(component_id = 'factorselect', component_property = 'value'))

def factorgraph(factor_input):
    df = factor_categories.query(f"factor_id == '{factor_input}'")

    fig = px.bar(df, x = 'factor_cat', y = 'avg_job_satisfaction',
                    labels = {'factor_cat':'Factor Category',
                             'avg_job_satisfaction':'Average Job Satisfaction'},
                    height = 600, width = 600,
                    color = 'factor_cat',
                    #symbol = 'chamber',
                    #opacity = 0.5,
                    hover_data = ['avg_job_satisfaction'])
    return fig
```
